## matplotlib line chart

### line chart 그리기 및 스타일링

경기도에서 서울로 전입한 사람 수를 line chart로 그려보면서 matplotlib의 기본 옵션들을 익혀보겠습니다. 먼저 plt.plot() 함수에 series를 전달하면 series의 index를 x축, series의 값들을 y축으로 하는 선 그래프를 그릴 수 있습니다. 그 외 옵션들은 다음과 같은 역할을 수행합니다.

- plt.figure: 도표 크기 지정
- plt.title: 도표 제목 지정
- plt.xticks: 도표 x축 각 틱별 라벨 지정
- plt.xlabel, plt.ylabel: x축 y축을 각각 설명하는 라벨 지정
- plt.legend: 도표 오른쪽 위에 선그래프를 설명하는 라벨 추가. 이 때, plot 함수 호출시에 라벨을 지정해주어야 함.

```python
from matplotlib import pyplot as plt

plt.style.use("ggplot")
plt.rc("font", family="Malgun Gothic")
plt.figure(figsize=(10,4))
plt.title("서울 -> 경기 인구 이동")
plt.xlabel("기간", size=9)
plt.ylabel("이동 인구수", size= 9)
plt.xticks(rotation=90, size=8)             # 대부분의 옵션은 plot 위에
plt.plot(df_seoul.loc["경기도"], label="서울 -> 경기", color= 'purple')  
plt.legend()                                # 레전드는 plot 아래에 들어가야 하는 옵션
```

### 여러 그래프를 한 도표 안에 그리기
여러 데이터에 대해서 plt.plot을 호출하면 한 도표에 여러 그래프를 그릴 수 있습니다 

```python

plt.rc("font", family="Malgun Gothic")
plt.figure(figsize=(10,4))
plt.title("서울 -> 경기 인구 이동")
plt.xlabel("기간", size=9)
plt.ylabel("이동 인구수", size= 9)
plt.xticks(rotation=90, size=8)           
plt.plot(df_seoul.loc["경기도"], label="서울 -> 경기")  
plt.plot(df_seoul.loc["경상북도"], label="서울 -> 경상북도") 
plt.legend() 
```

```python
destinations = [
    "부산광역시", "대구광역시", "인천광역시", "광주광역시", "대전광역시", "울산광역시", "세종특별자치시"
]
```

```python

plt.rc("font", family="Malgun Gothic")
plt.figure(figsize=(10,4))
plt.title("서울 -> 광역시 인구 이동")
plt.xlabel("기간", size=9)
plt.ylabel("이동 인구수", size= 9)
plt.xticks(rotation=90, size=8)           
for destination in destinations:
    plt.plot(df_seoul.loc[destination], label=f"서울 -> {destination}", marker = "o", markersize=3)
plt.legend() 
```

### 화면 분할
matplotlib에서 여러개의 도표를 한번에 만들고 싶을 때가 있다. 그럴 때는 figure의 subplot 기능을 사용하여 axe 객체들을 만들고, 각 axe별로 그래프를 그려주면 됩니다.

```python
fig = plt.figure(figsize=(12,4))

ax1 = fig.add_subplot(1,2,1)   # 행은 하나 열은 2개, 1번째 axe, 그래프
ax2 = fig.add_subplot(1,2,2)   # 행은 하나 열은 2개, 2번째 axe,그래프

ax1.set_xticklabels(df_seoul.loc["경기도"].index, rotation= 90, size=5)
ax2.set_xticklabels(df_seoul.loc["인천광역시"].index, rotation= 90, size=5)

ax1.plot(df_seoul.loc["경기도"], label= f"서울 -> 경기도", color='orange')
ax1.legend()
ax2.plot(df_seoul.loc["인천광역시"], label= f"서울 -> 인천", color='green')
ax2.legend
```

```python
fig = plt.figure(figsize=(12,8))

ax1 = fig.add_subplot(2,2,1)   
ax2 = fig.add_subplot(2,2,2)  
ax3 = fig.add_subplot(2,2,3) 
ax4 = fig.add_subplot(2,2,4) 

ax1.set_xticklabels(df_seoul.loc["경기도"].index, rotation= 90, size=5)
ax2.set_xticklabels(df_seoul.loc["인천광역시"].index, rotation= 90, size=5)
ax3.set_xticklabels(df_seoul.loc["대전광역시"].index, rotation= 90, size=5)
ax4.set_xticklabels(df_seoul.loc["대구광역시"].index, rotation= 90, size=5)

ax1.plot(df_seoul.loc["경기도"], label= f"서울 -> 경기도", color='orange')
ax1.legend()
ax2.plot(df_seoul.loc["인천광역시"], label= f"서울 -> 인천", color='green')
ax2.legend()
ax3.plot(df_seoul.loc["대전광역시"], label= f"서울 -> 대전광역시", color='blue')
ax3.legend()
ax4.plot(df_seoul.loc["대구광역시"], label= f"서울 -> 대구광역시", color='red')
ax4.legend()
```
